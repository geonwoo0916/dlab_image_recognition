```python
from sklearn import svm, metrics, model_selection
import joblib
import glob
from PIL import Image
import pandas as pd
import numpy as np

def 이미지_인식기( file_path ) :

  이미지수 = glob.glob(file_path + '*.png')
  이미지리스트 = []

  for i in 이미지수 :
    이미지 = Image.open(i).convert('L').resize((픽셀수, 픽셀수), Image.ANTIALIAS)
    이미지배열 = np.array( 이미지.getdata() ).reshape(픽셀수, 픽셀수)

    이미지리스트.append( 이미지배열.reshape(1, -1)[0] )

  이미지라벨 = [] #

  for i in range( len(이미지수) ) :
    if 'rhino' in 이미지수[i] :
      이미지라벨.append( 1 )
    else :
      이미지라벨.append( 2 )

  이미지데이터_df = pd.DataFrame( 이미지리스트 )
  이미지데이터_df['species[1R/2T]'] = 이미지라벨
  이미지데이터_df_name_array = 이미지데이터_df.columns.tolist()

  이미지클래스 = svm.SVC(C=1.0, kernel='rbf')

  이미지_x = 이미지데이터_df.iloc[:, 0:(18*18)]
  이미지_y = 이미지데이터_df['species[1R/2T]']

  이미지클래스.fit( 이미지_x, 이미지_y )

  이미지_예측 = 이미지클래스.predict( 이미지_x )
  정확도 = metrics.accuracy_score(이미지_y, 이미지_예측)
  print( "acc : {:.1f}".format( 정확도 * 100 )  )

  joblib.dump(이미지클래스, '/content/gbrive/MyDrive/Colab Notebooks/호랑이와코뿔소/tiger_rhino.pkl')

  print( "Complete Your Job")
```
